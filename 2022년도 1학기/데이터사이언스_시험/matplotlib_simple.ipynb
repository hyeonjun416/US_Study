### Line Plots
```python
plt.plot(x, y)
```
선 색의 표현
+ `'blue'` : 색 이름
+ `'g'` : short color code (rgbcmyk)
+ `'0.75'` : Grayscale 0~1
+ `'#FFDD44'` : Hexcode RRGGBB 00~FF
+ `'(1.0, 0.2, 0.3)'` : (R, G, B) 튜플 0~1
+ `'chartreuse'` : HTML 색 이름

선 스타일의 표현
+ linestyle='`solid`' : 직선 (-)
+ linestyle='`dashed`' : 점선 (--)
+ linestyle='`dashdot`' : 점과 선 (-.)
+ linestyle='`dotted`' : 점 (:)

````python
zero=np.zeros(1000)
zero
for style in ['-k', '--b', '-.g', ':r']: #2개 한번에 표현 가능
    plt.plot(x, zero, style, label="style : %s"%(style))
    zero+=1
plt.axis([0,2*np.pi+2, -1, 4]);
plt.legend()
````

### Scatter Plots
+ 방법 1 : ``plt.plot``/``ax.plot`` 에서 marker 문자 추가하기
```python
plt.plot(x, y, type, color=, markersize=, linewidth=, markerfacecolor=, markeredgecolor=, markeredgewidth=)
#x, y, 타입, 선색, 마커크기, 선굵기, 마커색, 마커테두리색, 마커테두리굵기
```
+ 방법 2 : ``plt.scatter()``
```python
plt.scatter(x, y, c=, s=, alpha=, cmap=)
#x, y, 색상, 사이즈, 투명도, 컬러맵
```

````python
x,y=1,1
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd', 'p', 'h']:
    plt.plot(x,y,marker, label="marker : %c"%(marker))
    x+=1 ; y+=1
plt.axis([0,20,0,15])
plt.legend(loc='right');
````

````python
x=np.random.random(100)
y=np.random.random(100)
colors=np.random.random(100)
sizes=1000*np.random.random(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar();
````

### Multiple Subplots
```python
plt.axes([left,bottom,width,height]) #위치, 크기 0~1 / axes 새로 만들기
plt.subplots(row, column, number) #row*column number개 axes 만들기 (grid)
plt.subplots_adjust(hspace=, wspace=) #간격조절
grid=plt.GridSpec(row, column, wspace=, hspace=) #row*column개 axes, 간격조절
    plt.subplot(grid[0,0]) #공간 할당
    plt.subplot(grid[0, 1:]) #공간 할당
```

````python
ax1=plt.axes()
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])
````

````python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
           fontsize=18, ha='center')
````

````python
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
````

### Relplot (Relational)
```python
sns.relplot(x=, y=, hue=, data=, kind=, col=, ci=)
#x, y, 카테고리, 데이터, 선/점, col을 통한 multi plot, 오차표현
#오차가 있을 경우 default : bootstrapping에 의한 95% confidence interval, 'sd'로 설정시 표준편차

````python
import seaborn as sns
sns.set_theme(style='darkgrid')
tips=sns.load_dataset("tips")
tips.head(5)
````

````python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter')
````

````python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', hue='smoker')
````

````python
sns.relplot(x='total_bill', y='tip', data=tips, kind='line')
````

````python
sns.relplot(x="total_bill", y="tip", hue="smoker", col="time", data=tips);
````

### Displot (Distributions)
```python
sns.displot(data, x=, col=, binwidth=, bins=, hue=, stat=, kind=)
#데이터, x, col을 통한 multiplot, bin굵기, bin갯수, 카테고리, 스탯, 종류
#스탯 : "density" : normalized(면적의 합이 1), "probability" : y값의 합이 1
#종류 : default는 히스토그램, "kde" : Kernel density estimation, "ecd" : Empirical cumulative distributions
```

````python
penguins=sns.load_dataset('penguins')
penguins.head()
````

````python
sns.displot(penguins, x='flipper_length_mm', bins=20, binwidth=3);
````

````python
sns.displot(penguins, x='flipper_length_mm', bins=20, binwidth=3, hue='species');
````

````python
sns.displot(penguins, x='flipper_length_mm', bins=20, binwidth=3, col='species');
````

````python
sns.displot(penguins, x='flipper_length_mm', kind='kde');
````

````python
sns.displot(penguins, x='flipper_length_mm', kind='ecdf');
````

````python
sns.displot(penguins, x='flipper_length_mm', y="bill_depth_mm"); #2차원
````

````python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm");
````

### Regplot, Lmplot

````python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', col='time'); #95% confidence interval
````

````python
sns.lmplot(x="size", y="tip", data=tips, x_jitter=.05); #add noise in the case of discrete independent variable
````

````python
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean);
````

````python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg");
````

````python
sns.pairplot(tips, x_vars=["total_bill", "size"], y_vars=["tip"],
             height=5, aspect=.8, kind="reg");
````
